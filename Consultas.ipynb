## 6/ Sacar el max y min

```python
pmax = data["popularity"].max()
pmin = data["popularity"].min()

print("Popularidad máxima:", pmax)
print("Popularidad mínima:", pmin)
```

## 7/ Ver a que franquicia, pelicula le pertenece la maxima popularidad

```python
# max-indice
imaxpopular= data["popularity"].idxmax()

# belongs_to_collection relacion a popularity
belongs_to_collection_max_popularity = data.loc[imaxpopular, "belongs_to_collection"]

print("Popularidad máxima:", pmax)
print("franquicia y/o pelicula al que pertenece la máxima a la popularidad :", belongs_to_collection_max_popularity)
```

## 13/ peliculas con mayor recaudacion

```python
# ordenar mayor -> menor
top_10_mas_recaudacion = data.nlargest(10, "revenue")

# ordenar menor -> mayor
top_10_menos_recaudacion = data.nsmallest(10, "revenue")

print("Top 10 películas con mayor recaudación:")
print(top_10_mas_recaudacion[["belongs_to_collection", "revenue"]])

print("\nTop 10 películas con menor recaudación:")
print(top_10_menos_recaudacion[["belongs_to_collection", "revenue"]])
```

## 14/ pelicula con mas recuadacion

```python
# != de 0
data_con_recaudacion = data[data["revenue"] != 0]

#mayor->menor y la pelicula con mas recaudacion
top_10_mas_recaudacion = data_con_recaudacion.nlargest(10, "revenue")

# top 10 peliculas + recaudaciones
print("Top 10 películas con mayor recaudación:")
print(top_10_mas_recaudacion[["title", "revenue"]])

#peliculas =! a 0
data_con_recaudacion_diferente_cero = data_con_recaudacion[data_con_recaudacion["revenue"] != 0]

#menor->mayor  pelicuala con menor recaudación
top_10_menos_recaudacion = data_con_recaudacion_diferente_cero.nsmallest(10, "revenue")

# verificar si existe las 10 peliculas =! 0
print("\nTop 10 películas con menor recaudación (diferente de 0):")
print(top_10_menos_recaudacion[["title", "revenue"]])
```
